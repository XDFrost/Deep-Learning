```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x = df['cgpa'],y =  df['resume_score'], hue=df['placed'])
```

```python
x = df.iloc[:,0:2]          # cgpa, resume score
y = df.iloc[:,-1]           # Output
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(x,y)
```

- y = w1x1 + w2x2 + b

```python
p.coef_             # Gives w1, w2 in this case
```

```python
p.intercept_            # gives b
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values, y.values, clf=p)        

# x.values give all values in x 
# y.values give all values in y
```

```python
p.get_params(deep=True)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(activation = 'logistic', max_iter=1000)
```

```python
mlp.fit(x,y)
```

```python
mlp.coefs_          # Gives w1, w2 in this case
```

```python
plot_decision_regions(x.values, y.values, clf=mlp)        
```
